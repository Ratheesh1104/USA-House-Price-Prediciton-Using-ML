```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder,StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_squared_error
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from xgboost import XGBRegressor
```

```python
df = pd.read_csv("usa_housing_kaggle.csv")
df. head()
```

# Feature Engineering #

```python
df["HouseAge"] = 2025 - df["YearBuilt"]
df=df.drop(["YearBuilt"],axis=1)
```

```python
df['Price_m2'] = df['Price']/ df['SquareFeet']
```

```python
df['Price_per_bedrooms'] = df['Price']/df['Bedrooms']
df['Age_lot_interaction'] = df['HouseAge']*df['LotSize']
```

```python
df['Crime_School_Interaction'] = df['CrimeRate']*df['SchoolRating']
df = df.drop(columns=['CrimeRate','SchoolRating','LotSize','HouseAge'])
```

```python
df = df.drop(['ZipCode'],axis=1)
```

```python
corr = df.corr()
sns.heatmap(corr , annot = True , cmap = 'coolwarm' , linewidths = 0.1,fmt = '.2f')
```

# Outlier Detection #

```python
for col in df:
    plt.figure(figsize = (10,5))
    sns.boxplot(x=df[col])
    plt.title(f'{col} Boxplot',fontsize=14)
    plt.xlabel(col,fontsize = 12)
    plt.grid(True , linestyle ='--',alpha = 0.7)
    plt.show()
```

```python
sns.scatterplot(x='Bathrooms',y='Bedrooms',hue = 'Price',  data = df)
```

```python
x = df.drop(['Price'],axis =1)
y = df['Price']

x_train , x_test , y_train , y_test = train_test_split(x,y,test_size = 0.2 , random_state = 42)
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.fit_transform(x_test)
```

```python
df.head()
```

# Linear Regression Model #

```python
lr = LinearRegression()
lr.fit(x_train_scaled,y_train)
predict = lr.predict(x_test_scaled)
score = lr.score(x_test_scaled , y_test)
print(f'Score :',score)
```

# Random Forest Regressor Model #

```python
rf = RandomForestRegressor(n_estimators =100, random_state = 42)
rf.fit(x_train_scaled ,y_train)
rf_predict = rf.predict(x_test_scaled)
rf_mse = mean_squared_error(y_test,rf_predict)
rf_r2 = r2_score(y_test , rf_predict)

print("MSE : " , rf_mse)
print("R2 : ",rf_r2)
```

# XGBoost Model #

```python
xgb_model = XGBRegressor(n_estimators=100, random_state=42)
xgb_model.fit(x_train_scaled, y_train)
xgb_pred = xgb_model.predict(x_test_scaled)
xgb_mse = mean_squared_error(y_test, xgb_pred)
xgb_r2 = r2_score(y_test, xgb_pred)

print("XGBoost - MSE:", xgb_mse)
print("XGBoost - R:", xgb_r2)
```

# Model Score Compare #

```python
model_score = {
    'Random Forest' : rf_r2 * 100,
    'XGBootst': xgb_r2 * 100,
    'Linear Regression': score * 100
    
}

plt.figure(figsize = (10,5))

plt.bar(model_score.keys() , model_score.values(), color = ['red','blue','green'])
plt.title("Model Compare" , fontsize = 12)
plt.xlabel("Model" , fontsize=12)
plt.ylabel("R2 Score %" , fontsize=12)

plt.show()
```

**Here " Random Forest " Model Give better r2 Score than Other two Model , So We took Random Forest model for Price Prediciton**
